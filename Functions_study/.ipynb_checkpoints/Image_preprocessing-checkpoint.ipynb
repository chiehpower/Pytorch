```python
import torchvision.transforms as T
import PIL.Image as Image
import cv2
import torchvision.transforms
from matplotlib import pyplot as plt
import torch
%matplotlib inline
```

```python
# Read the image
image_path = "../img/00000001.jpg"
image=Image.open(image_path)
print("image.size:{} \nimage.format:{} \nimage.mode:{}".format(image.size, image.format, image.mode))
plt.imshow(image)
```

# CenterCrop

```python
# Generate a CenterCrop class object, used to crop the picture from the center to 224 * 224
crop_obj = torchvision.transforms.CenterCrop((224, 224))  
image2 = crop_obj(image) 
```

```python
# Save the cropped picture
image2.save("../img/00000001_after.jpg", format='JPEG')
```

```python
plt.imshow(image2)
print("image.size:{} \nimage.format:{} \nimage.mode:{}".format(image2.size, image2.format, image2.mode))
```

# Resize and CenterCrop

```python
transform = T.Compose(
    [
                T.ToTensor(),
                T.ToPILImage(),
                T.Resize((107,107)),
                T.CenterCrop(107),
    ]
)
# Use only Resize(107)
transform_cc = T.Compose(
    [

        T.ToTensor(),
        T.ToPILImage(),
        T.Resize(107),
        T.CenterCrop(107),
    ]
)
```

```python
image3 = transform(image) 
plt.imshow(image3)
print("image.size:{} \nimage.format:{} \nimage.mode:{}".format(image3.size, image3.format, image3.mode))
```

We can see if you only use resize(107), the picture will be zoomed out. The upper one is more better.

```python
image4 = transform_cc(image) 
plt.imshow(image4)
print("image.size:{} \nimage.format:{} \nimage.mode:{}".format(image4.size, image4.format, image4.mode))
```

# Use cv2

Using `cv2.read` is different with `Image.open`. The orders of ToPILImage() and ToTensor() are different.

```python
transform_piltotensor = T.Compose(
    [
                T.ToPILImage(),
                T.Resize((107,107)),
                T.CenterCrop(107),
                T.ToTensor(),
    ]
)
```

```python
img = cv2.imread(image_path)
print(img.shape)
img2 = transform_piltotensor(img) 
print(img2.size())
```

```python
img3 = img2.unsqueeze(0).numpy().ravel()
img3
```
